```python
# import dependencies
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
# create dataframe from file
file_path = '../Resources/austinHousingData.csv'
a_sales_df = pd.read_csv(file_path)

a_sales_df.head()
```

```python
HouseInfo_df = a_sales_df[['zpid', 'description', 
                           'propertyTaxRate', 'hasAssociation', 'hasCooling', 'hasGarage', 'hasHeating', 
                           'hasSpa', 'hasView', 'homeType', 'yearBuilt', 'latestPrice', 'latest_saleyear', 
                           'lotSizeSqFt', 'livingAreaSqFt', 'numOfBathrooms', 'numOfBedrooms', 'numOfStories']]

HouseFeatures_df = a_sales_df[['zpid', 'garageSpaces', 'parkingSpaces', 'numOfAccessibilityFeatures', 'numOfAppliances', 
                               'numOfParkingFeatures', 'numOfPatioAndPorchFeatures', 'numOfSecurityFeatures', 
                               'numOfWaterfrontFeatures', 'numOfWindowFeatures', 'numOfCommunityFeatures']]

HouseLoc_df = a_sales_df[['zpid', 'city', 'streetAddress', 'zipcode', 'latitude', 'longitude']]

HouseSchool_df = a_sales_df[['zpid', 'numOfPrimarySchools', 'numOfElementarySchools', 'numOfMiddleSchools', 
                             'numOfHighSchools', 'avgSchoolDistance', 'avgSchoolRating', 'avgSchoolSize',  
                             'MedianStudentsPerTeacher']]
```

```python
HouseInfo_df.to_csv('../Resources/Database_CSVs/Austin_House_Info.csv', index=False)
```

```python
HouseLoc_df.to_csv('../Resources/Database_CSVs/Austin_House_Location.csv', index=False)
```

```python
HouseFeatures_df.to_csv('../Resources/Database_CSVs/Austin_House_Features.csv', index=False)
```

```python
HouseSchool_df.to_csv('../Resources/Database_CSVs/Austin_House_Schools.csv', index=False)
```
